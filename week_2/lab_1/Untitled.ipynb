```python
import pandas as pd
```

```python
df = pd.read_csv("Spacex.csv")
df.head()
```

```python
dates = df['Date']
new_dates = []
for date in dates:
    l = date.split('-')
    new_dates.append(l[2] + '-' + l[1] + '-' + l[0])
```

```python
new_dates
```

```python
df['Date'] = new_dates
```

```python
df.head()
```

```python
df.to_csv('new_spacex.csv', index=False)
```
